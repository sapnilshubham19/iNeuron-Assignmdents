```python
df = pd.read_csv('C:\\Users\\sapni\\Downloads\\iNeuron\\Assignments\\titanic-train.csv')
df
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df1 = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
df1.Sex = df1.Sex.map({'male' : 1 , 'female' : 2})
```

```python
X_train , X_test , y_train , y_test = train_test_split(df1,target,test_size=0.2)
```

```python
model.score(X_test,y_test)
```
